```python
def palindrome(s):
  s1=s[::-1]
  if s1==s:
    print("palindrome")
  else:
    print("not")

palindrome("nitin")
```

```python
def palindrome(s):
  count=0
  for i in range(len(s)//2):
    if s[i]==s[len(s)-1-i]:
      count+=1
  if count==len(s)//2:
    print("palindrome")
  else:
    print("not")
```

```python
s1="abbc"
s2="cbab"
d1={}
d2={}
for i in s1:
  if i not in d1:
    d1[i]=1
  else:
    d1[i]=d1[i]+1
for i in s2:
  if i not in d2:
    d2[i]=1
  else:
    d2[i]=d2[i]+1

print(d1,d2)
if d1==d2:
  print("equal")
else:
  print("not equal")
```
